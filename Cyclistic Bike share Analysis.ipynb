```python
# Importing required libraries

import pandas as pd
import numpy as np
import datetime
```

```python
# Reading CSV files of Oct 2021 to June 2022
file_1 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202110-divvy-tripdata.csv')
file_2 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202111-divvy-tripdata.csv')
file_3 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202112-divvy-tripdata.csv')
file_4 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202201-divvy-tripdata.csv')
file_5 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202202-divvy-tripdata.csv')
file_6 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202203-divvy-tripdata.csv')
file_7 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202204-divvy-tripdata.csv')
file_8 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202205-divvy-tripdata.csv')
file_9 = pd.read_csv('C:/Users/Nandkumar/Desktop/Cyclistic_data/202206-divvy-tripdata.csv')
```

```python
# merging all dataframes into one dataframe
data_frames = [file_1, file_2, file_3, file_4, file_5, file_6, file_7, file_8, file_9]
merged_df = pd.concat(data_frames)

merged_df.head()
```

```python
merged_df.info()
```

```python
# Converting object data type to datetime 
merged_df['started_at'] = pd.to_datetime(merged_df['started_at'])
merged_df['ended_at'] = pd.to_datetime(merged_df['ended_at'])
```

```python
# Calculating Ride length 
merged_df['ride_length'] = merged_df['ended_at'] - merged_df['started_at']
```

```python
merged_df.head()
```

```python
# Splitting start date and time from column started_at 
merged_df[['start_date','start_time']] = merged_df['started_at'].astype(str).str.split(' ',n=1,expand=True) 
merged_df.info()
```

```python
# Converting object data type to datetime 
merged_df['start_time'] = pd.to_datetime(merged_df['start_time'])
merged_df['start_date'] = pd.to_datetime(merged_df['start_date'])
```

```python
# Finding weekday of the date
merged_df['week_day'] = merged_df['start_date'].apply(lambda x:x.weekday())
merged_df['week_day'].unique()
```

```python
# Converting weekday values into weekday
days_dict = {0:'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'}
merged_df['week_day'] = merged_df['week_day'].apply(lambda y:days_dict[y])
```

```python
# Sorting dataframe by started_at in ascending order
merged_df.sort_values(by=['started_at'], inplace=True, ascending=True)
merged_df.head()
```

```python
# Checking null values 
merged_df.isnull().sum()
```

### Null values are less as compare to our dataset, there is no need to drop any of the null rows 

```python
# Checking duplicate values
merged_df.duplicated().any()
```

```python
# Extracting year and month from start_date column
merged_df['year'] = pd.DatetimeIndex(merged_df['start_date']).year
merged_df['month'] = pd.DatetimeIndex(merged_df['start_date']).month
```
